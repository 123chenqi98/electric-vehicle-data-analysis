# 🚗 新能源汽车数据探索

本Notebook用于探索Kaggle Electric Vehicle Population Data数据集

```python
# 导入必要的库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 导入自定义模块
import sys
sys.path.append('../src')
from data_processing import load_data, get_basic_info
```

```python
# 加载数据
df = load_data('../data/raw/electric_vehicle_population_data.csv')

# 查看数据基本信息
info = get_basic_info(df)

print(f"数据形状: {info['shape']}")
print(f"\n列名: {info['columns']}")
print(f"\n数据类型:\n{pd.DataFrame.from_dict(info['dtypes'], orient='index', columns=['dtype'])}")
print(f"\n缺失值统计:\n{pd.DataFrame.from_dict(info['missing_count'], orient='index', columns=['count'])}")
print(f"\n缺失值百分比:\n{pd.DataFrame.from_dict(info['missing_percentage'], orient='index', columns=['percentage'])}")
```

```python
# 查看数据前几行
df.head()
```

```python
# 数值特征统计描述
df.describe().round(2)
```

```python
# 分类特征统计
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\n{col}:")
    print(df[col].value_counts().head(10))
```

```python
# 可视化：价格分布
plt.figure(figsize=(12, 6))
sns.histplot(df['Base MSRP'].dropna(), bins=50, kde=True)
plt.title('车辆价格分布')
plt.xlabel('价格 ($)')
plt.ylabel('数量')
plt.show()
```

```python
# 可视化：续航里程分布
plt.figure(figsize=(12, 6))
sns.histplot(df['Electric Range'].dropna(), bins=50, kde=True)
plt.title('续航里程分布')
plt.xlabel('续航里程 (英里)')
plt.ylabel('数量')
plt.show()
```

```python
# 可视化：车型年份分布
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Model Year', order=df['Model Year'].value_counts().index[:10])
plt.title('车型年份分布')
plt.xlabel('年份')
plt.ylabel('数量')
plt.xticks(rotation=45)
plt.show()
```

```python
# 可视化：品牌分布（Top 10）
top_brands = df['Make'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_brands.values, y=top_brands.index)
plt.title('品牌分布（Top 10）')
plt.xlabel('数量')
plt.ylabel('品牌')
plt.show()
```

```python
# 可视化：价格与续航的关系
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Electric Range', y='Base MSRP', alpha=0.6)
plt.title('价格与续航里程的关系')
plt.xlabel('续航里程 (英里)')
plt.ylabel('价格 ($)')
plt.show()
```

```python
# 相关性分析
numeric_cols = df.select_dtypes(include=[np.number]).columns
correlation_matrix = df[numeric_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('特征相关性热力图')
plt.show()
```
